```python
characters = demographics.dropna(subset=['Gender'])
```

```python
character_total = characters.loc[:, ['Name', 'Gender']]
character_total = character_total.drop_duplicates()
character_count = character_total.count()[0]

#Create data frame to display the total number of unique players
pd.DataFrame({'Character Count': [character_count]})
```

```python
#Obtain value counts of new characters each year and calculate their percentages
year_totals = demographics['Year'].value_counts()
year_percent = (year_totals / character_count) * 100

#Create a summary data frame to display the new character counts and percentages
new_character_stats = pd.DataFrame({'TotalNewCharacters': year_totals, 'PercentofTotalCharacters': year_percent})

#Round the percentages to two decimal places
new_character_stats = new_character_stats.round(2)

#Give the data for Percentage of Characters a cleaner formatting
new_character_stats['PercentofTotalCharacters'] = new_character_stats['PercentofTotalCharacters'].map('{:,.2f}%'.format)

#Display the data frame
new_character_stats
```

```python
first_appearances = new_character_stats.reset_index()
```

```python
first = first_appearances.rename(columns = {'index': 'Year'})
```

```python
#Obtain value counts of genders and calculate their percentages
gender_totals = demographics['Gender'].value_counts()
gender_percent = (gender_totals / character_count) * 100

#Create a summary data frame to display the gender counts and percentages
gender_demographics_df = pd.DataFrame({'Total Count': gender_totals, 
                                       'Percentage of Characters': gender_percent})

#Round the percentages to two decimal places
gender_demographics_df = gender_demographics_df.round(2)

#Give the data for Percentage of Players a cleaner formatting
gender_demographics_df['Percentage of Characters'] = gender_demographics_df['Percentage of Characters'].map('{:,.2f}%'.format)

#Display the data frame
gender_demographics_df
```

```python
gender_pie = gender_demographics_df.rename(columns = {'index': 'Gender'})
gender_pie
```

```python
good = demographics[demographics['Alignment'] == 'Good']
bad = demographics[demographics['Alignment'] == 'Bad']
neutral = demographics[demographics['Alignment'] == 'Neutral']
```

```python
good_count = good.groupby(['Gender']).count()['Alignment'].rename('Good Count')
good_percent = (good_count / 12940) * 100
bad_count = bad.groupby(['Gender']).count()['Alignment'].rename('Bad Count')
bad_percent = (bad_count / 12940) * 100
neutral_count = neutral.groupby(['Gender']).count()['Alignment'].rename('Neutral Count')
neutral_percent = (neutral_count / 12940) * 100
total_count = good_count + bad_count + neutral_count
total_percent = (total_count / 12940) * 100
```

```python
gender_alignment = pd.DataFrame({'Good': good_count, 'Good Percent': good_percent, 'Bad': bad_count, 
                                 'Bad Percent': bad_percent, 'Neutral': neutral_count, 
                                 'Neutral Percent': neutral_percent, 
                                 'Total': total_count, 'Total Percent': total_percent})
gender_alignment.dropna(inplace=True)
```

```python
#Round the percentages to two decimal places
gender_alignment = gender_alignment.round(2)

#Give the data for Percentage of Characters a cleaner formatting
gender_alignment['Good Percent'] = gender_alignment['Good Percent'].map('{:,.2f}%'.format)
gender_alignment['Bad Percent'] = gender_alignment['Bad Percent'].map('{:,.2f}%'.format)
gender_alignment['Neutral Percent'] = gender_alignment['Neutral Percent'].map('{:,.2f}%'.format)
gender_alignment['Total Percent'] = gender_alignment['Total Percent'].map('{:,.2f}%'.format)

#Display completed gender/alignment dataframe
gender_alignment
```

```python
bar1 = gender_alignment.drop(columns=['Good','Bad','Neutral','Total'])
```
